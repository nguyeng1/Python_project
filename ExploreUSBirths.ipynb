```python
def read_csv(filename):
    splitted_data = open(filename).read().split("\n")
    string_list = splitted_data[1:(len(splitted_data)-1)]
    final_list = []
    for row in string_list:
        int_fields= []
        string_fields = row.split(",")
        for field in string_fields:
            int_field = int(field)
            int_fields.append(int_field)
        final_list.append(int_fields)
    return(final_list)
cdc_list=read_csv("US_births_1994-2003_CDC_NCHS.csv")
print(cdc_list[2000:2250])
```

```python
def month_births(list_of_lists):
    births_per_month = {}
    for item in list_of_lists:
        month_value=item[1]
        births_value=item[4]
        if month_value in births_per_month:
            births_per_month[month_value]= births_per_month[month_value]+births_value
        else:
            births_per_month[month_value]=births_value
    return(births_per_month)
cdc_month_births = month_births(cdc_list)
cdc_month_births
```

```python
def dow_births(list_of_lists):
    births_per_week = {}
    for item in list_of_lists:
        weekday_value = item[3]
        births_value = item[4]
        if weekday_value in births_per_week:
            births_per_week[weekday_value] = births_per_week[weekday_value]+births_value
        else:
            births_per_week[weekday_value] = births_value
    return(births_per_week)
cdc_day_births = dow_births(cdc_list)
cdc_day_births
```

```python
def calc_counts(data,column):
    if column==1:
        year_births = {}
        for item in data:
            if item[0] in year_births:
                year_births[item[0]] = year_births[item[0]]+item[4]
            else:
                year_births[item[0]] = item[4]
        return(year_births)
    if column==2:
        month_births = {}
        for item in data:
            if item[1] in month_births:
                month_births[item[1]] = month_births[item[1]]+item[4]
            else:
                month_births[item[1]] = item[4]
        return(month_births)
    if column==3:
        dom_births = {}
        for item in data:
            if item[2] in dom_births:
                dom_births[item[2]] = dom_births[item[2]]+item[4]
            else:
                dom_births[item[2]] = item[4]
        return(dom_births)
        
    if column==4:
        dow_births = {}
        for item in data:
            if item[3] in dow_births:
                dow_births[item[3]] = dow_births[item[3]]+item[4]
            else:
                dow_births[item[3]] = item[4]
        return(dow_births)

cdc_year_births = calc_counts(cdc_list,1)
cdc_month_births = calc_counts(cdc_list,2)
cdc_dom_births = calc_counts(cdc_list,3)
cdc_dow_births = calc_counts(cdc_list,4)
print(cdc_year_births)
print(cdc_month_births)
print(cdc_dom_births)
print(cdc_dow_births)
```

```python
def min_max(any_dictionary):
    max_value = max(any_dictionary.values())
    min_value = min(any_dictionary.values())
    return(max_value,min_value)
print(min_max(cdc_year_births))
print(min_max(cdc_month_births))
print(min_max(cdc_dom_births))
print(min_max(cdc_dow_births))
```

```python
#Find the difference in consecutive values between year 1994-2003
from collections import OrderedDict
def value_difference(list_of_lists,dow):
    #First calculate total value of yearly dow (index 3) and append results in a list
    yearly_dow = OrderedDict()
    if dow==1:
        for item in list_of_lists:
            if item[3]==1:
                if item[0] in yearly_dow:
                    yearly_dow[item[0]]=yearly_dow[item[0]]+item[4]
                else:
                    yearly_dow[item[0]]=item[4]
    print(yearly_dow)
     #Secondly find the differences in consecutive values   
    yearly_differences = []
    for key in yearly_dow:
        if key+1<=2003:
            difference = yearly_dow[key]-yearly_dow[key+1]
            yearly_differences.append(difference)
    return(yearly_differences)

cdc_yearly_differences = value_difference(cdc_list,1)
print(cdc_yearly_differences)
```
